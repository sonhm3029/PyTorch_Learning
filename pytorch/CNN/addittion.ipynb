# Một số kiến thức cơ bản về ảnh:

## 1. Ảnh màu:

Là ma trận 3 chiều các điểm ảnh. Trong đó có 3 channel màu R, G, B, mỗi channel đóng vai trò là 1 ma trận 2 chiều các điểm anh.

![](../../images/rgb_img.png)

## 2. Ảnh xám:

Là ma trận 2 chiều các điểm ảnh, trong đó giá trị của các điểm ảnh nằm trong đoạn [0-255].

## 3. Chuyển hệ màu của ảnh

Mỗi pixel trong ảnh màu được biểu diễn bằng 3 giá trị (r, g, b) còn trong ảnh xám chỉ cần 1 giá trị x để biểu diễn/

Khi chuyển từ ảnh màu sang ảnh xám, ta có thể dùng công thức: 

`x = r x 0.299 + g x 0.587 + b x 0.114`

Tuy nhiên khi chuyển ngược lại, bạn chỉ biết giá trị x và cần đi tìm r,g,b nên sẽ không chính xác.

### Padding

Như ở trên  thì mỗi lần thực hiện phép convolution xong thì kích thước ma trận Y đều nhỏ hơn X. Tuy nhiên giờ ta muốn ma trận Y thu được có kích thước bằng ma trận X thì ta có một cách đó là thêm padding ( Thêm các điểm ảnh có giá trị 0 vào viền ngoài ma trận X).

![](../../images/convolution3.png)

Vậy bây giờ ta đã thu được ma trận Y có kích thước bằng ma trận X ban đầu, đây còn gọi là phép convolution với `padding = 1`. `Padding = k` nghĩa là thêm `k` vector 0 vào mỗi phía của ma trận.

### Stride

Như ở trên ta thực hiện tuần tự các phần tử trong ma trận X, thu được ma trận Y cùng kích thước ma trận X, ta gọi là stride=1.

![](../../images/convolution4.png)

Tuy nhiên nếu stride=k (k > 1) thì ta chỉ thực hiện phép tính convolution trên các phần tử x<sub>1+i * k,1+j * k</sub>
