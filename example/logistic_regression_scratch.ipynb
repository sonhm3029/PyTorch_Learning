```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
import numpy.random as random


breastCancer = datasets.load_breast_cancer()
X = breastCancer.data
y = breastCancer.target

n_samples, n_features = X.shape

# DF represent the data in table
# dfBefore = {key: X[:, idx] for idx, key in enumerate(breastCancer.feature_names)}
# dfBefore['result'] = [breastCancer.target_names[i] for i in y]
df = pd.DataFrame(X, columns=breastCancer.feature_names)
print(df)
# pd.plotting.scatter_matrix(df, c=y, figsize=(30,30), 
#                             marker='o', hist_kwds={'bins':20}, s=10,
#                             alpha=.8)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2910)

```

```python

class LogisticRegression:
    def __init__(self, lr, max_iters):
        self.lr = lr
        self.max_iters = max_iters
    def fit(self, X, y):
        n_samples, n_features = X.shape
        X = X.T
        y = y.reshape(-1,n_samples)
        w = random.rand(n_features,1)
        print(w.shape)
        bias = 1
        # Data nomaliztion:
        
        
        # Training loop
        for epoch in range(self.max_iters):
            y_linear = w.T.dot(X)
            y_pred = self.active_func(y_linear)
            loss = (y_pred - y).dot(X.T)
            # update
            w -= self.lr*loss.T
        self.w = w   

    def predict(self, X):
        print(self.w)
        y_linear = self.w.T.dot(X.T)
        print(y_linear)
        y_pred = self.active_func(y_linear)
        result = [1 if value >0.5 else 0 for value in y_pred[0]]
        return result

    def active_func(self, X):
        return 1/ (1 + np.exp(-X))


if __name__ == "__main__":
    model = LogisticRegression(lr=0.0001, max_iters=1000)
    model.fit(X_train, y_train)
    y_predict = model.predict(X_test)
    print(f"accuracy : {accuracy_score(y_pred=y_predict, y_true=y_test)}")
    
```
