K-means clustering là thuật toán cơ bản nhất của học không giám sát. Thuật toán được sử dụng với tập dữ liệu chỉ có các điểm dữ liệu mà không cho trước các nhãn của chúng. Công việc của thuật toán này là phân các điểm dữ liệu đó thành các nhóm dữ liệu sao cho các điểm dữ liệu trong cùng một nhóm thì có tính chất giống nhau.

![](../../images/k-means-0.png)

Thuật toán:

- Tập dữ liệu **`X`**:

![](../../images/k-means-1.png)

Trong đó **`xi`** thuộc ma trận **`X`** là các vector. K là số các cluster mà ta muốn phân ra

- Tập nhãn **`Y`**:

![](../../images/k-means-2.png)

Trong đó **`yi`** thuộc ma trận **`Y`** là các vector nhãn dán với ![](../../images/k-means-3.png) là nhãn của **`xi`**. Qua đó, nếu **`xi`** được phân vào cluster `k` thì y<sub>ik</sub> = 1 và y<sub>ij</sub> = 0

![](../../images/k-means-4.png)

- Tập center **`M`**:

![](../../images/k-means-5.png)

Trong đó **`mi`** thuộc ma trận **`M`** là các vector chỉ center của các cluster, có số chiều bằng với **`xi`**


**Hàm mất mát:**

Đối với mỗi điểm dữ liệu ta cần tìm sao cho khoảng cách từ điểm dữ liệu đó đến center của cluster nào gần nó nhất hay giá trị sau đây nhỏ nhất:

![](../../images/k-means-7.png)

do tính chất y<sub>ik</sub> = 1 và y<sub>ij</sub> = 0 nên ta có thể viết hàm mất mát trên toàn bộ điểm dữ liệu như sau:

![](../../images/k-means-8.png)



Cần tìm **`Y`** và **`M`** để cho hàm mất mát đạt giá trị min kí hiệu như sau:

![](../../images/k-means-6.png)


Tóm tắt thuật toán như sau:

Đầu vào: Tập dữ liệu **`X`** và số lượng cluster cần tìm là **`K`**
Đầu ra: Các center **`M`** và label cho từng điểm dữ liệu **`Y`**

1. Chọn **`K`** điểm bất kì từ tập dữ liệu **`X`** làm các center ban đầu
2. Tính toán khoảng cách từ các điểm thuộc tập **`X`** đến các center của các cluster và so sánh xem điểm gần center nào hơn thì update label cho điểm đó vào cluster tương ứng
3. Nếu như center không thay đổi giữa 2 lần loop thì dừng loop
4. Update các center của các cluster bằng cách lấy trung bình cộng của tất cả các điểm dữ liệu trong cluster từ bước 2.
5. Quay lại bước 2.
